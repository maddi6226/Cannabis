```python
# Program to measure the similarity between  
# two sentences using cosine similarity. 
from nltk.corpus import stopwords 
from nltk.tokenize import word_tokenize 
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer(language='english')
def comparesentences(X,Y): 
    # X = input("Enter first string: ").lower() 
# Y = input("Enter second string: ").lower() 
#X ="I love horror movies"
#Y ="Lights out is a horror movie"
  
# tokenization
    symbols = "!\"#'$%&()*+-,./:;<=>’?@[\]^_`{|}~\n"
    for i in symbols:
        X= X.replace(i," ")
        Y= Y.replace(i," ")
        X=X.lower()
        Y=Y.lower()
    X_list = word_tokenize(X)  
    Y_list = word_tokenize(Y) 
  

    sw = stopwords.words('english')  
    l1 =[];l2 =[] 
  

    X_set = {w for w in X_list if not w in sw}  
    Y_set = {w for w in Y_list if not w in sw} 
  
    rvector = X_set.union(Y_set)  
    for w in rvector: 
        if w in X_set: 
            l1.append(1) # create a vector 
        else:
            l1.append(0) 
        if w in Y_set:
            l2.append(1) 
        else:
            l2.append(0) 
    c = 0
  
 
    for i in range(len(rvector)): 
        c+= l1[i]*l2[i] 
    cosine = c / float((sum(l1)*sum(l2))**0.5) 
    return cosine

```

```python
R=float(input("Enter The minimum rating that you want:"))
M=str(input("Enter the type of Medical_useage: "))
E=str(input("Enter the type of effects: "))
o=int(input("how many suggestions you want? : "))
test=data.loc[(data['Rating'] >= R) &(data['Medical_useage']==M)]
test=test.sort_values('Rating', ascending=False)

test.insert(7, "CosineSimilarity_Score", "Any")
for i in range(len(test)):
    test.iloc[i,7]=comparesentences(test.iloc[i,4],E) 

test=test.sort_values('CosineSimilarity_Score',ascending=False)
test=test[test['CosineSimilarity_Score']>0] 
test1=test.head(o)
test1[["Strain","Type","Effects","Rating","CosineSimilarity_Score","Medical_useage"]]
```
